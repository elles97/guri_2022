```python
# 1) 필요한 라이브러리 불어오기
import matplotlib.pyplot as plt
import csv
# 2) 기온 데이터 불어오기
f=open('seoul_temp.csv',encoding='cp949')
data=csv.reader(f)
next(data)
mx_temp=[]
for row in data:
    if row[-1]!='':
    #if row[-1]!='' and row[0].split('-')[1]=='10' and row[0].split('-')[2]=='01':    
       mx_temp.append(float(row[-1])) 
    
# 3) 기온 데이터에서 최고 기온을 순서대로 저장하기
# 4) 저장된 데이터를 다양한그래프로 그리기
plt.plot(mx_temp)
# 5) 결과 그래프 보여주기
plt.show()
```
